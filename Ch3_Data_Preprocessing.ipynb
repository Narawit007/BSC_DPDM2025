#Data Cleaning

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

## Missing Data

### Drop Record ที่มี missing ออก

drop ถ้ามี missing

```python
df_cleaned = df.dropna()
print("DataFrame after dropping rows with missing values:")
from IPython.display import display
display(df_cleaned)
```

### dropเฉพาะเกิดmissingในตัวแปรที่เราจะใช้

```python
columns_to_check = ['dfo_country', 'dfo_ended', 'dfo_other_country', 'dfo_began', 'dfo_dead', 'dfo_cause', 'dfo_severity']
df_filtered = df.dropna(subset=columns_to_check)
print("DataFrame after dropping rows with missing values in specified columns:")
from IPython.display import display
display(df_filtered)
```

### เติม Dummy

```python
df_filled_dummies = df.copy()
df_filled_dummies['dfo_other_country'] = df_filled_dummies['dfo_other_country'].fillna('unknown')
df_filled_dummies = pd.get_dummies(df_filled_dummies, columns=['dfo_other_country'], prefix='dfo_other_country', drop_first=True)
print("DataFrame after filling missing 'dfo_other_country' with 'unknown' and creating dummy variables:")
from IPython.display import display
display(df_filled_dummies.head())
```

### เติมmissing ด้วยค่ากลาง

```python
mode_dfo_cause = df['dfo_cause'].mode()[0]
df['dfo_cause'].fillna(mode_dfo_cause, inplace=True)
print(f"Missing values in 'dfo_cause' filled with mode: {mode_dfo_cause}")
print("Number of missing values in 'dfo_cause' after imputation:")
print(df['dfo_cause'].isnull().sum())

from IPython.display import display
display(df.head())
```
