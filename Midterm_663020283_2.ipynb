````python
import pandas as pd
import os

# Define the base output directory for filtered files
base_output_dir = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Output - Filtered Daily Rain'

# Ensure valid_nodes is available from unique_nodes
if 'unique_nodes' not in locals() or unique_nodes is None:
    print("Error: 'unique_nodes' ‡πÑ‡∏°‡πà‡∏û‡∏ö. ‡πÇ‡∏õ‡∏£‡∏î‡∏£‡∏±‡∏ô‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÇ‡∏´‡∏•‡∏î Node ‡∏à‡∏≤‡∏Å‡πÑ‡∏ü‡∏•‡πå Master.")
    # Fallback in case unique_nodes was somehow not set, though it should be.
    try:
        master_file_path = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Input - (Operation Daily rain from API_ 2025)/Stations in the model_HII.csv'
        df_master_stations = pd.read_csv(master_file_path)
        if 'Node' in df_master_stations.columns:
            valid_nodes = set(df_master_stations['Node'].unique())
            print("valid_nodes ‡∏ñ‡∏π‡∏Å‡πÇ‡∏´‡∏•‡∏î‡∏à‡∏≤‡∏Å‡πÑ‡∏ü‡∏•‡πå Master ‡πÅ‡∏•‡πâ‡∏ß (‡πÄ‡∏õ‡πá‡∏ô fallback).")
        else:
            raise ValueError(f"‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'Node' ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå Master: {master_file_path}")
    except Exception as e:
        print(f"Error: ‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡πÇ‡∏´‡∏•‡∏î valid_nodes (fallback): {e}")
        valid_nodes = set()
else:
    valid_nodes = set(unique_nodes) # Use the existing unique_nodes

# Ensure monthly_directories is available
if 'monthly_directories' not in locals():
    print("Error: 'monthly_directories' ‡πÑ‡∏°‡πà‡∏û‡∏ö. ‡πÇ‡∏õ‡∏£‡∏î‡∏£‡∏±‡∏ô‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÑ‡∏î‡πÄ‡∏£‡∏Å‡∏ó‡∏≠‡∏£‡∏µ‡∏£‡∏≤‡∏¢‡πÄ‡∏î‡∏∑‡∏≠‡∏ô.")
    # Fallback to define monthly_directories if not found
    base_input_dir = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Input - (Operation Daily rain from API_ 2025)'
    monthly_directories = [
        os.path.join(base_input_dir, '2025-07'),
        os.path.join(base_input_dir, '2025-08'),
        os.path.join(base_input_dir, '2025-09'),
        os.path.join(base_input_dir, '2025-10'),
        os.path.join(base_input_dir, '2025-11'),
        os.path.join(base_input_dir, '2025-12')
    ]
    print("monthly_directories ‡∏ñ‡∏π‡∏Å‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡∏°‡πà‡πÅ‡∏•‡πâ‡∏ß (‡πÄ‡∏õ‡πá‡∏ô fallback)." + "\n")

station_code_column = 'tele_station_oldcode'

print("‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÅ‡∏•‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏´‡∏°‡πà‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏≤‡∏¢‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏Å‡∏£‡∏≠‡∏á‡πÅ‡∏•‡πâ‡∏ß...")
print(f"‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà‡∏Å‡∏£‡∏≠‡∏á‡πÅ‡∏•‡πâ‡∏ß‡∏à‡∏∞‡∏ñ‡∏π‡∏Å‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÉ‡∏ô '{base_output_dir}'")
print("-" * 80)

total_files_processed = 0
total_stations_before_filter_sum = 0 # Sum of unique stations across all files before filter
total_stations_after_filter_sum = 0  # Sum of unique stations across all files after filter

for monthly_dir_path in monthly_directories:
    folder_name = os.path.basename(monthly_dir_path)
    output_monthly_dir = os.path.join(base_output_dir, folder_name)
    os.makedirs(output_monthly_dir, exist_ok=True) # Create output directory if it doesn't exist

    if not os.path.isdir(monthly_dir_path):
        print(f"  Warning: {monthly_dir_path} ‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡πÑ‡∏î‡πÄ‡∏£‡∏Å‡∏ó‡∏≠‡∏£‡∏µ‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á ‡∏Ç‡πâ‡∏≤‡∏°‡πÑ‡∏õ.")
        continue

    csv_files = [f for f in os.listdir(monthly_dir_path) if f.endswith('.csv')]

    if not csv_files:
        print(f"  ‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÑ‡∏ü‡∏•‡πå CSV ‡∏û‡∏ö‡πÉ‡∏ô {folder_name}. ‡∏Ç‡πâ‡∏≤‡∏°‡πÑ‡∏õ.")
        continue

    files_in_month_processed = 0
    stations_before_month_avg_accumulator = []
    stations_after_month_avg_accumulator = []

    for filename in csv_files:
        file_path = os.path.join(monthly_dir_path, filename)
        output_file_path = os.path.join(output_monthly_dir, filename)

        try:
            df_daily = pd.read_csv(file_path)

            if station_code_column not in df_daily.columns:
                print(f"  Warning: ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå '{station_code_column}' ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå {filename}. ‡∏Ç‡πâ‡∏≤‡∏°‡πÑ‡∏ü‡∏•‡πå‡∏ô‡∏µ‡πâ.")
                continue

            df_daily[station_code_column] = df_daily[station_code_column].astype(str)

            # Count unique stations before filtering
            stations_before_this_file = df_daily[station_code_column].nunique()

            # Filter the DataFrame
            df_filtered = df_daily[df_daily[station_code_column].isin(valid_nodes)].copy()

            # Count unique stations after filtering
            stations_after_this_file = df_filtered[station_code_column].nunique()

            # Save the filtered DataFrame to the new output path
            df_filtered.to_csv(output_file_path, index=False)

            stations_before_month_avg_accumulator.append(stations_before_this_file)
            stations_after_month_avg_accumulator.append(stations_after_this_file)
            files_in_month_processed += 1
            total_files_processed += 1

        except FileNotFoundError:
            print(f"  Error: ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà‡∏£‡∏∞‡∏ö‡∏∏: {file_path}. ‡∏Ç‡πâ‡∏≤‡∏°‡πÑ‡∏õ.")
        except Exception as e:
            print(f"  Error: ‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÑ‡∏ü‡∏•‡πå {file_path}: {e}. ‡∏Ç‡πâ‡∏≤‡∏°‡πÑ‡∏õ.")

    if files_in_month_processed > 0:
        avg_before = int(sum(stations_before_month_avg_accumulator) / files_in_month_processed)
        avg_after = int(sum(stations_after_month_avg_accumulator) / files_in_month_processed)
        print(f"‡πÄ‡∏î‡∏∑‡∏≠‡∏ô {folder_name}: ‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏Å‡πà‡∏≠‡∏ô‡∏Å‡∏£‡∏≠‡∏á: {avg_before}, ‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á: {avg_after} ({files_in_month_processed} ‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•)")
        total_stations_before_filter_sum += sum(stations_before_month_avg_accumulator)
        total_stations_after_filter_sum += sum(stations_after_month_avg_accumulator)
    else:
        print(f"‡πÄ‡∏î‡∏∑‡∏≠‡∏ô {folder_name}: ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÑ‡∏î‡πâ.")

print("-" * 80)
print(f"‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô‡∏Å‡∏≤‡∏£‡∏î‡∏≥‡πÄ‡∏ô‡∏¥‡∏ô‡∏Å‡∏≤‡∏£. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏´‡∏°‡πà‡∏ó‡∏µ‡πà‡∏Å‡∏£‡∏≠‡∏á‡πÅ‡∏•‡πâ‡∏ß‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î {total_files_processed} ‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏ô '{base_output_dir}'")
if total_files_processed > 0:
    print(f"‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏£‡∏ß‡∏°‡∏Ç‡∏≠‡∏á‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏Å‡πà‡∏≠‡∏ô‡∏Å‡∏£‡∏≠‡∏á‡∏ï‡πà‡∏≠‡∏ß‡∏±‡∏ô: {total_stations_before_filter_sum / total_files_processed:.0f}")
    print(f"‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏£‡∏ß‡∏°‡∏Ç‡∏≠‡∏á‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á‡∏ï‡πà‡∏≠‡∏ß‡∏±‡∏ô: {total_stations_after_filter_sum / total_files_processed:.0f}")
else:
    print("‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏î‡∏ñ‡∏π‡∏Å‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•.")
````

````python
import pandas as pd
import os
from calendar import monthrange

# --- ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡πÄ‡∏™‡πâ‡∏ô‡∏ó‡∏≤‡∏á‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏Å‡∏≤‡∏£‡∏Å‡∏£‡∏≠‡∏á‡πÅ‡∏•‡πâ‡∏ß ---
output_filtered_path = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Output - Filtered Daily Rain'
target_col = 'rainfall_value'
station_col = 'tele_station_oldcode'

# ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå‡πÄ‡∏î‡∏∑‡∏≠‡∏ô
months = ['2025-07', '2025-08', '2025-09', '2025-10', '2025-11', '2025-12']

for month_str in months:
    dir_path = os.path.join(output_filtered_path, month_str)
    if not os.path.isdir(dir_path):
        continue

    print(f"\nüìä [ ‡∏™‡∏£‡∏∏‡∏õ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏π‡∏ç‡∏´‡∏≤‡∏¢‡∏£‡∏≤‡∏¢‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ: {month_str} ]")
    print(f"{'Station ID':<15} | {'Expected':<10} | {'Missing':<10} | {'Missing %'}")
    print("-" * 55)

    # 1. ‡∏´‡∏≤‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡πÉ‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ô‡∏±‡πâ‡∏ô (‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤ 1 ‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏Ñ‡∏ß‡∏£‡∏°‡∏µ‡∏Å‡∏µ‡πà Record)
    year, month = map(int, month_str.split('-'))
    days_in_month = monthrange(year, month)[1]

    # 2. ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏∏‡∏Å‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ô‡∏±‡πâ‡∏ô‡πÄ‡∏Ç‡πâ‡∏≤‡∏î‡πâ‡∏ß‡∏¢‡∏Å‡∏±‡∏ô
    all_data = []
    csv_files = [f for f in os.listdir(dir_path) if f.endswith('.csv')]

    for f in csv_files:
        try:
            temp_df = pd.read_csv(os.path.join(dir_path, f))
            all_data.append(temp_df[[station_col, target_col]])
        except:
            continue

    if not all_data:
        print("‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ô‡∏µ‡πâ")
        continue

    df_month = pd.concat(all_data)

    # 3. ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏£‡∏≤‡∏¢‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ
    # ‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô Record ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏à‡∏£‡∏¥‡∏á (‡πÑ‡∏°‡πà‡∏ô‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô NaN)
    station_stats = df_month.groupby(station_col)[target_col].count().reset_index()
    station_stats.columns = [station_col, 'actual_count']

    # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏™‡πà‡∏ß‡∏ô‡∏ï‡πà‡∏≤‡∏á
    station_stats['missing_count'] = days_in_month - station_stats['actual_count']
    # ‡∏Å‡∏£‡∏ì‡∏µ‡∏ö‡∏≤‡∏á‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏™‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡πâ‡∏≥‡πÄ‡∏Å‡∏¥‡∏ô‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ß‡∏±‡∏ô ‡πÉ‡∏´‡πâ‡πÅ‡∏Å‡πâ‡∏Ñ‡πà‡∏≤‡∏ï‡∏¥‡∏î‡∏•‡∏ö‡πÄ‡∏õ‡πá‡∏ô 0
    station_stats.loc[station_stats['missing_count'] < 0, 'missing_count'] = 0

    station_stats['missing_percent'] = (station_stats['missing_count'] / days_in_month) * 100

    # 4. ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏• (‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÄ‡∏â‡∏û‡∏≤‡∏∞ 10 ‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡πÅ‡∏£‡∏Å‡∏ó‡∏µ‡πà‡∏´‡∏≤‡∏¢‡πÄ‡∏¢‡∏≠‡∏∞‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡∏°‡∏≤‡πÇ‡∏ä‡∏ß‡πå ‡∏´‡∏£‡∏∑‡∏≠‡πÇ‡∏ä‡∏ß‡πå‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î)
    # ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏à‡∏≤‡∏Å‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏ó‡∏µ‡πà‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏≤‡∏¢‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
    station_stats = station_stats.sort_values(by='missing_percent', ascending=False)

    for index, row in station_stats.iterrows():
        print(f"{row[station_col]:<15} | {days_in_month:<10} | {int(row['missing_count']):<10} | {row['missing_percent']:>8.2f}%")

    # ‡∏™‡∏£‡∏∏‡∏õ‡∏†‡∏≤‡∏û‡∏£‡∏ß‡∏°‡πÄ‡∏î‡∏∑‡∏≠‡∏ô
    avg_missing = station_stats['missing_percent'].mean()
    print(f"\n>> ‡∏†‡∏≤‡∏û‡∏£‡∏ß‡∏°‡πÄ‡∏î‡∏∑‡∏≠‡∏ô {month_str}: ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏≤‡∏¢‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ {avg_missing:.2f}% ‡∏à‡∏≤‡∏Å‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î {len(station_stats)} ‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ")
````

````python
import pandas as pd
import os
from calendar import monthrange

# --- ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤ Path ---
input_dir = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Output - Filtered Daily Rain'
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå‡πÉ‡∏´‡∏°‡πà‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏Å‡πá‡∏ö‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡πÄ‡∏Å‡∏ì‡∏ë‡πå (‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏≤‡∏¢‡πÑ‡∏°‡πà‡πÄ‡∏Å‡∏¥‡∏ô 20%)
final_output_dir = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Output - Clean Final Data'

target_col = 'rainfall_value'
station_col = 'tele_station_oldcode'
threshold = 20.0  # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÄ‡∏Å‡∏ì‡∏ë‡πå‡πÄ‡∏õ‡∏≠‡∏£‡πå‡πÄ‡∏ã‡πá‡∏ô‡∏ï‡πå‡∏ó‡∏µ‡πà‡∏¢‡∏≠‡∏°‡πÉ‡∏´‡πâ‡∏´‡∏≤‡∏¢‡πÑ‡∏î‡πâ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î

if not os.path.exists(final_output_dir):
    os.makedirs(final_output_dir)

months = ['2025-07', '2025-08', '2025-09', '2025-10', '2025-11', '2025-12']

for month_str in months:
    src_month_path = os.path.join(input_dir, month_str)
    dest_month_path = os.path.join(final_output_dir, month_str)

    if not os.path.isdir(src_month_path):
        continue

    if not os.path.exists(dest_month_path):
        os.makedirs(dest_month_path)

    print(f"\nüîç ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡πÄ‡∏î‡∏∑‡∏≠‡∏ô: {month_str}")

    # 1. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏´‡∏≤‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏Ñ‡∏ß‡∏£‡∏°‡∏µ‡πÉ‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ô‡∏±‡πâ‡∏ô
    year, month = map(int, month_str.split('-'))
    days_in_month = monthrange(year, month)[1]

    # 2. ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏£‡∏∞‡∏ö‡∏∏‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏ï‡∏±‡∏î‡∏≠‡∏≠‡∏Å
    csv_files = [f for f in os.listdir(src_month_path) if f.endswith('.csv')]
    temp_list = []
    for f in csv_files:
        temp_list.append(pd.read_csv(os.path.join(src_month_path, f)))

    df_all_month = pd.concat(temp_list)

    # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡πÄ‡∏õ‡∏≠‡∏£‡πå‡πÄ‡∏ã‡πá‡∏ô‡∏ï‡πå‡∏´‡∏≤‡∏¢‡∏£‡∏≤‡∏¢‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ
    station_stats = df_all_month.groupby(station_col)[target_col].count().reset_index()
    station_stats['missing_percent'] = (1 - (station_stats[target_col] / days_in_month)) * 100

    # ‡∏Ñ‡∏±‡∏î‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡πÄ‡∏Å‡∏ì‡∏ë‡πå (Missing <= 20%)
    valid_stations = station_stats[station_stats['missing_percent'] <= threshold][station_col].tolist()
    dropped_stations = station_stats[station_stats['missing_percent'] > threshold][station_col].tolist()

    print(f"   ‚úÖ ‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡πÄ‡∏Å‡∏ì‡∏ë‡πå: {len(valid_stations)} ‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ")
    print(f"   ‚ùå ‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡∏±‡∏î‡∏≠‡∏≠‡∏Å (>20%): {len(dropped_stations)} ‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ")
    if dropped_stations:
        print(f"      ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠: {', '.join(dropped_stations)}")

    # 3. ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏´‡∏°‡πà‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡πÄ‡∏Å‡∏ì‡∏ë‡πå
    for f in csv_files:
        df_daily = pd.read_csv(os.path.join(src_month_path, f))
        # ‡∏Å‡∏£‡∏≠‡∏á‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡∏ó‡∏µ‡πà‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô‡∏•‡∏¥‡∏™‡∏ï‡πå valid_stations
        df_clean = df_daily[df_daily[station_col].isin(valid_stations)]

        # ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏•‡∏á‡πÉ‡∏ô‡∏ó‡∏µ‡πà‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏´‡∏°‡πà
        df_clean.to_csv(os.path.join(dest_month_path, f), index=False)

print("\n‚ú® ‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô! ‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà‡∏Å‡∏£‡∏≠‡∏á‡πÅ‡∏•‡πâ‡∏ß‡∏≠‡∏¢‡∏π‡πà‡∏ó‡∏µ‡πà‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå: Output - Clean Final Data")
````

````python
import pandas as pd
import os

# --- 1. ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡πÑ‡∏ü‡∏•‡πå‡πÅ‡∏•‡∏∞‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå ---
historical_file = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Daily rain - (Statistics 2012-2024)/hii_daily_rain.csv'
clean_data_dir = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Output - Clean Final Data'
imputed_output_dir = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Output - Final Imputed Data'

# ‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå (Historical)
hist_station_col = 'station_name'
hist_data_col = 'data'
hist_date_col = 'date' # *** ‡∏£‡∏ö‡∏Å‡∏ß‡∏ô‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏ô‡∏µ‡πâ‡∏≠‡∏µ‡∏Å‡∏Ñ‡∏£‡∏±‡πâ‡∏á (‡πÄ‡∏ä‡πà‡∏ô date ‡∏´‡∏£‡∏∑‡∏≠ rainfall_datetime) ***

# ‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô (2025)
curr_station_col = 'tele_station_oldcode'
curr_data_col = 'rainfall_value'
curr_date_col = 'rainfall_datetime'

if not os.path.exists(imputed_output_dir):
    os.makedirs(imputed_output_dir)

# --- 2. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤ Median ‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå ---
print("‚è≥ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÇ‡∏´‡∏•‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå‡πÅ‡∏•‡∏∞‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤ Median...")
try:
    # ‡∏≠‡πà‡∏≤‡∏ô‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡∏£‡∏∞‡∏´‡∏¢‡∏±‡∏î‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏Ñ‡∏ß‡∏≤‡∏°‡∏à‡∏≥
    df_hist = pd.read_csv(historical_file)

    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà (‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà 'date' ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏ä‡∏∑‡πà‡∏≠‡πÉ‡∏ô‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ hist_date_col)
    if hist_date_col not in df_hist.columns:
        # ‡∏•‡∏≠‡∏á‡∏´‡∏≤‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡∏≥‡∏ß‡πà‡∏≤ 'date' ‡∏´‡∏£‡∏∑‡∏≠ 'time'
        possible_date_cols = [c for c in df_hist.columns if 'date' in c.lower() or 'time' in c.lower()]
        if possible_date_cols:
            hist_date_col = possible_date_cols[0]
            print(f"üîé ‡∏û‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå‡∏ä‡∏∑‡πà‡∏≠: {hist_date_col}")

    df_hist[hist_date_col] = pd.to_datetime(df_hist[hist_date_col])
    df_hist['day_of_year'] = df_hist[hist_date_col].dt.dayofyear

    # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Median ‡πÅ‡∏¢‡∏Å‡∏ï‡∏≤‡∏°‡∏ä‡∏∑‡πà‡∏≠‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡πÅ‡∏•‡∏∞‡∏ß‡∏±‡∏ô‡πÉ‡∏ô‡∏£‡∏≠‡∏ö‡∏õ‡∏µ
    median_map = df_hist.groupby([hist_station_col, 'day_of_year'])[hist_data_col].median().reset_index()
    median_map.columns = [hist_station_col, 'day_of_year', 'median_val']
    print(f"‚úÖ ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Median ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à (‡∏û‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• {len(median_map)} ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£)")
except Exception as e:
    print(f"‚ùå ‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÑ‡∏ü‡∏•‡πå‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå: {e}")
    median_map = None

# --- 3. ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Å‡∏£‡∏∞‡∏ö‡∏ß‡∏ô‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏¥‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• (Imputation) ---
if median_map is not None:
    months = ['2025-07', '2025-08', '2025-09', '2025-10', '2025-11', '2025-12']

    for month_str in months:
        src_path = os.path.join(clean_data_dir, month_str)
        dest_path = os.path.join(imputed_output_dir, month_str)

        if not os.path.isdir(src_path): continue
        if not os.path.exists(dest_path): os.makedirs(dest_path)

        print(f"üîπ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏ï‡∏¥‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏î‡∏∑‡∏≠‡∏ô: {month_str}")
        csv_files = [f for f in os.listdir(src_path) if f.endswith('.csv')]

        for f in csv_files:
            df = pd.read_csv(os.path.join(src_path, f))
            df[curr_date_col] = pd.to_datetime(df[curr_date_col])
            df['day_of_year'] = df[curr_date_col].dt.dayofyear
            df[curr_station_col] = df[curr_station_col].astype(str)

            # ‡∏ô‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏ß‡πà‡∏≤‡∏á‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡∏ï‡∏¥‡∏°
            if df[curr_data_col].isna().any():
                # ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÇ‡∏î‡∏¢‡∏à‡∏±‡∏ö‡∏Ñ‡∏π‡πà curr_station_col (2025) ‡πÄ‡∏Ç‡πâ‡∏≤‡∏Å‡∏±‡∏ö hist_station_col (Historical)
                df = df.merge(median_map,
                              left_on=[curr_station_col, 'day_of_year'],
                              right_on=[hist_station_col, 'day_of_year'],
                              how='left')

                # ‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤‡∏ß‡πà‡∏≤‡∏á‡∏î‡πâ‡∏ß‡∏¢ Median
                df[curr_data_col] = df[curr_data_col].fillna(df['median_val'])

                # ‡∏´‡∏≤‡∏Å‡∏¢‡∏±‡∏á‡∏ß‡πà‡∏≤‡∏á (‡∏Å‡∏£‡∏ì‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ß‡∏±‡∏ô‡∏ô‡∏±‡πâ‡∏ô) ‡πÉ‡∏´‡πâ‡πÄ‡∏ï‡∏¥‡∏° 0
                df[curr_data_col] = df[curr_data_col].fillna(0)

                # ‡∏•‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏™‡πà‡∏ß‡∏ô‡πÄ‡∏Å‡∏¥‡∏ô
                df.drop(columns=[hist_station_col, 'median_val', 'day_of_year'], inplace=True)
            else:
                df.drop(columns=['day_of_year'], inplace=True)

            # ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå‡πÅ‡∏•‡πâ‡∏ß
            df.to_csv(os.path.join(dest_path, f), index=False)

    print(f"\n‚ú® ‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢! ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå‡∏ñ‡∏π‡∏Å‡πÄ‡∏Å‡πá‡∏ö‡πÑ‡∏ß‡πâ‡∏ó‡∏µ‡πà: {imputed_output_dir}")
````

#‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà4 ‡∏£‡∏ß‡∏°‡∏õ‡∏£‡∏¥‡∏°‡∏≤‡∏ì‡∏ù‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ß‡∏±‡∏ô‡πÄ‡∏õ‡πá‡∏ô‡∏ú‡∏•‡∏£‡∏≤‡∏¢‡πÄ‡∏î‡∏∑‡∏≠‡∏ô

````python
import pandas as pd
import os

# --- 1. ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå ---
# ‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå‡∏ó‡∏µ‡πà‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡∏∂‡πà‡∏á‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤ Median ‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß‡∏à‡∏≤‡∏Å‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤
imputed_data_dir = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Output - Final Imputed Data'
output_final_file = '/content/drive/MyDrive/‡∏™‡∏≠‡∏ö3‡πÄ‡∏ó‡∏≠‡∏°2/Y_monthly_rain_HII_final_summary.csv'

# ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•
months_list = ['2025-07', '2025-08', '2025-09', '2025-10', '2025-11', '2025-12']

final_summary = []

print("‚è≥ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏ú‡∏•‡∏£‡∏ß‡∏°‡∏£‡∏≤‡∏¢‡πÄ‡∏î‡∏∑‡∏≠‡∏ô...")

for folder_name in months_list:
    folder_path = os.path.join(imputed_data_dir, folder_name)

    if not os.path.isdir(folder_path):
        print(f"‚ö†Ô∏è ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå: {folder_name} (‡∏Ç‡πâ‡∏≤‡∏°...)")
        continue

    # ‡πÅ‡∏¢‡∏Å‡∏õ‡∏µ‡πÅ‡∏•‡∏∞‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡∏à‡∏≤‡∏Å‡∏ä‡∏∑‡πà‡∏≠‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå
    year_val = int(folder_name.split('-')[0])
    month_val = int(folder_name.split('-')[1])

    print(f"üîπ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÄ‡∏î‡∏∑‡∏≠‡∏ô: {folder_name}")

    # ‡∏î‡∏∂‡∏á‡πÑ‡∏ü‡∏•‡πå CSV ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡πÉ‡∏ô‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå‡∏Ç‡∏≠‡∏á‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ô‡∏±‡πâ‡∏ô‡πÜ
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

    month_data_list = []
    for f in csv_files:
        try:
            temp_df = pd.read_csv(os.path.join(folder_path, f))
            # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô (‡∏£‡∏´‡∏±‡∏™‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ ‡πÅ‡∏•‡∏∞ ‡∏Ñ‡πà‡∏≤‡∏ù‡∏ô)
            if 'tele_station_oldcode' in temp_df.columns and 'rainfall_value' in temp_df.columns:
                month_data_list.append(temp_df[['tele_station_oldcode', 'rainfall_value']])
        except Exception as e:
            print(f"  ‚ùå ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå {f} ‡πÑ‡∏î‡πâ: {e}")

    if month_data_list:
        # ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏∏‡∏Å‡∏ß‡∏±‡∏ô‡∏Ç‡∏≠‡∏á‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ô‡∏±‡πâ‡∏ô‡πÄ‡∏Ç‡πâ‡∏≤‡∏î‡πâ‡∏ß‡∏¢‡∏Å‡∏±‡∏ô
        df_month = pd.concat(month_data_list)

        # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏ú‡∏•‡∏£‡∏ß‡∏°‡∏£‡∏≤‡∏¢‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡πÅ‡∏¢‡∏Å‡∏ï‡∏≤‡∏°‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ
        monthly_sum = df_month.groupby('tele_station_oldcode')['rainfall_value'].sum().reset_index()

        # ‡∏õ‡∏£‡∏±‡∏ö‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡πâ‡∏ï‡∏£‡∏á‡∏ï‡∏≤‡∏°‡πÑ‡∏ü‡∏•‡πå‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á
        monthly_sum['month'] = month_val
        monthly_sum['year'] = year_val
        monthly_sum.rename(columns={'tele_station_oldcode': 'Node', 'rainfall_value': 'rain'}, inplace=True)

        final_summary.append(monthly_sum)

# --- 2. ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏∏‡∏Å‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡πÅ‡∏•‡∏∞‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏ú‡∏• ---
if final_summary:
    df_final = pd.concat(final_summary)

    # ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏ï‡∏≤‡∏° Node ‡πÅ‡∏•‡∏∞ Month ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏ß‡∏¢‡∏á‡∏≤‡∏°‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á
    df_final = df_final.sort_values(by=['Node', 'year', 'month']).reset_index(drop=True)

    # ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÑ‡∏ü‡∏•‡πå CSV
    df_final.to_csv(output_final_file, index=False)

    print(f"\n‚ú® ‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î!")
    print(f"‚úÖ ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÑ‡∏ü‡∏•‡πå‡∏™‡∏£‡∏∏‡∏õ‡∏£‡∏≤‡∏¢‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ó‡∏µ‡πà: {output_final_file}")

    # ‡πÅ‡∏™‡∏î‡∏á‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• 5 ‡πÅ‡∏ñ‡∏ß‡πÅ‡∏£‡∏Å
    print("\n--- ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå ---")
    print(df_final.head())
else:
    print("‚ùå ‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÑ‡∏ü‡∏•‡πå‡∏™‡∏£‡∏∏‡∏õ")
````
